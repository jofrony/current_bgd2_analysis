```python
import matplotlib.pyplot as plt
import os
```

```python
tmp={"10Hz": 0.2,"5Hz": 0.41, "3Hz": 0.57}
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
csfont = {"fontname": "Arial",
          'fontsize': 16}
plt.title("Ratio decrease in active neurons due to 'stronger' P2 activation", **csfont)
for i, k in enumerate(tmp):
    plt.bar([i],[tmp[k] * 1e2], color="black")
plt.xticks([0,1,2],["10 Hz", "5 Hz", "3 Hz"])
ax.tick_params(axis='both', which='major', labelsize=22)
ax.tick_params(axis='both', which='minor', labelsize=22)
plt.ylim(0,100)
plt.ylabel(" Percentage of reduction in P1 (%)", **csfont)
plt.savefig(os.path.join("inhibited", "ratio_decrease.svg"))
```
