```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
csfont = {"fontname": "Arial",
      'fontsize': 22}
name = "SPN"
pos = np.arange(3)
experiment_names = ['control', '-80 mV reversal', "x2 SPN conductance"]
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks(pos, **csfont)
ax.set_xticklabels(experiment_names, rotation=45, **csfont)

ax.tick_params(axis='both', which='major', labelsize=22)
ax.tick_params(axis='both', which='minor', labelsize=22)
plt.ylabel("percentage inhibited plateaus (%)", **csfont)
plt.bar(pos,[50, 67.3,66.5], color="black")
plt.savefig(os.path.join("compare_plateau_simulations", f"percentage_inhibited_plateaus_{name}.svg"))
```
