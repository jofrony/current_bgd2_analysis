```python
import os
import h5py
from snudda import SnuddaInit
from snudda import SnuddaPlace
from snudda import SnuddaDetect
from snudda import SnuddaPrune
from snudda.input import SnuddaInput
import json
```

```python
os.environ["SNUDDA_DATA"] = os.environ["BASALGANGLIADATA"]
```

```python
seed=12345
```

```python
local = os.path.join("/home/jofrony/Documents/Repositories/current_bgd2/threshold/local")
network_path = os.path.join(local, "networks/striatum")

si = SnuddaInit(network_path=network_path, random_seed=seed)
si.define_striatum(num_iSPN=50, num_dSPN=50, num_FS=10, num_LTS=5,num_ChIN=1,volume_type="cube", neurons_dir="$DATA/neurons")
si.write_json()
```

```python
with open("/home/jofrony/Documents/Repositories/current_bgd2/threshold/local/networks/striatum/network-config.json") as f:
    di = json.load(f)
```

```python
with open("/home/jofrony/Documents/Repositories/current_bgd2/threshold/local/networks/striatum/above_threshold.json") as f:
    ki = json.load(f)
```

```python
for n in di["Neurons"]:
    
    if "dSPN" in n or "iSPN" in n:
        print(n)
        n_type = n.split("_")[0].lower()
        index = int(n.split("_")[1])
        #di["Neurons"][n].pop("morphology") 
        #di["Neurons"][n].pop("parameters") 
        di["Neurons"][n].update({"parameterKey": ki[n_type][index][0]})
        di["Neurons"][n].update({"morphologyKey": ki[n_type][index][1]})
```

```python
with open("/home/jofrony/Documents/Repositories/current_bgd2/threshold/local/networks/striatum/network-config.json", "w") as f:
    json.dump(di, f, indent=4)
```

```python
sp = SnuddaPlace(network_path=network_path)
sp.place()
sd = SnuddaDetect(network_path=network_path)
sd.detect()
sp = SnuddaPrune(network_path=network_path);
sp.prune()
```
