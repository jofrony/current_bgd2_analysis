```python
import os
import glob
import shutil
import json
```

```python
os.environ["BASALGANGLIADATA"]
```

```python
bgd = "/home/jofrony/Documents/Repositories/BasalGangliaData/data/neurons/striatum/*"
```

```python
with open("/home/jofrony/Documents/Repositories/current_bgd2/threshold/local/networks/striatum/above_threshold.json") as f:
    ki = json.load(f)
```

```python
for sources in glob.glob(bgd):
    for source in glob.glob(os.path.join(sources, "*")):
        
        if "dspn" in source or "ispn" in source:
            name = os.path.basename(source).split("-")[1]

            meta_tmp = dict()
            meta = json.load(open(os.path.join(source, "meta.json")))
            
            shutil.copy(os.path.join(source, "meta.json"), os.path.join(source, "meta_old.json"))
            for pair in ki[name]:
                
                p = pair[0]
                m = pair[1]
                
                if p in meta.keys() and m in meta[p].keys():
                
                    if p in meta_tmp:
                        meta_tmp[p].update({m: meta[p][m]})
                    else:
                        meta_tmp.update({p: {m: meta[p][m]}})

            with open(os.path.join(source, "meta.json"), "w") as f:
                json.dump(meta_tmp,f)
                    
        
```
